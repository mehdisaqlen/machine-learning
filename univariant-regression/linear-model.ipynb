# Linear Regression Model
A linear function based on single variable $x$ that predict the price of a house.

This linear function represents $f_{w,b}{(x)} = xw+b$. This function is a model that gives $y\hat{}$.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# X => hours of study
# Y => score in the exam
# Z => testing data set containing hours
X = np.array([1,2,3,5,7,9,12,15,16])
Y = np.array([45,52,75,79,85,88,95,97,99])
```

```python
reg_model = LinearRegression()
reg_model.fit(X.reshape(-1,1), Y.reshape(-1,1))
```

```python
# Prediction
y_pred = reg_model.predict(X.reshape(-1, 1))

# hour = 8
# y_pred = reg_model.predict([[hour]])
# score = y_pred[0][0]
# print(f"If you study {hour} hours/day, you may score {round(score,1)}%")
```

```python
# Scatter plot with regression line
plt.scatter(X, Y, label='Data points')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.title("Score Predictor")
plt.legend()
plt.show()
```
